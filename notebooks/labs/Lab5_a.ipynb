```r
unit <- function(x, b0, b1){
    h = b0 + b1 * x
    
    return (1/(1 + exp(-h)))
} 
unit_2input <- function(x1, x2, b0, b1, b2){
    h = b0 + b1 * x1 + b2 * x2
    
    return (1/(1 + exp(-h)))
} 
```

```r
data = tibble(x=seq(-10, 10, 0.1),
              y1=unit(x, 1, 1),
              y2=unit(x, 1, -1)
             )

p1 <- ggplot(data, aes(x, y1)) +
    geom_line(size=2, color='blue') + 
    geom_line(aes(x, y2, color='red'), size=2)

p2 <- ggplot(data, aes(x, y1)) + 
    geom_line(aes(x, y1 + y2), size=2)

grid.arrange(p1, p2, nrow=1)
```

```r
data = tibble(x=seq(-10, 10, 0.01),
              y11=unit(x, 1, 10),
              y22=unit(x, 1, -10)
             )

p1 <- ggplot(data, aes(x, y11)) +
    geom_line(size=2, color='red') + 
    geom_line(aes(x, y22), size=2, color='blue')
p2 <- ggplot(data, aes(x, y11 + y22)) +
    geom_line(size=2)
grid.arrange(p1, p2, nrow=1)
```

```r
data = tibble(x=seq(-10, 10, 0.01),
              y1=unit(x, 1, 1),
              y2=unit(x, 1, -1),
              y11=unit(x, 1, 10),
              y22=unit(x, 1, -10)
             )

p1 <- ggplot(data, aes(x, y1 + y2)) +
    geom_line(size=2, color='blue') + 
    geom_line(aes(x, y11 + y22), size=2, color='red')
p2 <- ggplot(data, aes(x, y1 + y2 + y11 + y22)) +
    geom_line(size=2)

grid.arrange(p1, p2, nrow=1)
```

```r
data = tibble(x=seq(-10, 10, 0.01),
              y1=unit(x, 1, 1),
              y2=unit(x, -3, -1),
              y11=unit_2input(y1, y2, .5, .7, 3),
              y22=unit_2input(y1, y2, 1, -.3, -1)
             )

p1 <- ggplot(data, aes(x,y11)) +
    geom_line(size=2, color='red') + 
    geom_line(aes(x, y22), size=2, color='blue')

p2 <- ggplot(data, aes(x,y11 + y22)) +
    geom_line(size=2)

grid.arrange(p1, p2, nrow=1)
```
